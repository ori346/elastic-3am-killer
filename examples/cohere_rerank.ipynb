```python
from typing import Union
import cohere
from cohere import Client, ClientV2

import sys, os, time, json

notebook_dir = os.getcwd()
parent_dir = os.path.dirname(notebook_dir)
if parent_dir not in sys.path:
    sys.path.insert(0, parent_dir)

import utilities.environment_config as config
```

```python
with open(config.document_file_location, 'r', encoding='utf-8') as f:
    movies = json.load(f)

extracts = list()
# We only process a slice of the documents since it would take too long to process all
# See the faster version of this in this notebook
for document in movies:
    if 'extract' in document:
        extracts.append(document['extract'])
```

```python
query = "What is a funny zombie or horror movie with a romantic twist?"
```

```python
def cohere_rerank(
    client: Union[Client, ClientV2], model: str, query: str, documents: list[str]
) -> dict:
    return client.rerank(model=model, query=query, documents=documents)
```
