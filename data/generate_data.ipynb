```python
import json
from faker import Faker

# Initialize Faker to generate data
fake = Faker()

def generate_movie_data(num_movies):
    """
    Generates a list of flat JSON objects, each representing a movie.
    """
    movies = []
    for i in range(num_movies):
        movie = {
            "movie_id": i + 1,
            "title": fake.sentence(nb_words=3).replace('.', ''),  # e.g., 'The Last Stand'
            "year": fake.year(),
            "genre": fake.word(),
            "rating": fake.random_int(min=1, max=10),
            "director": fake.name(),
            "description": fake.paragraph(nb_sentences=5),  # A single text field for the description
            "is_oscar_winner": fake.boolean()
        }
        movies.append(movie)
    return movies

# Generate 1000 movie records
movie_data = generate_movie_data(1000)

# Write the data to a JSON file
file_path = "movies.json"
with open(file_path, 'w') as f:
    json.dump(movie_data, f, indent=2)

print(f"Successfully generated {len(movie_data)} movie records and saved to {file_path}")

```
